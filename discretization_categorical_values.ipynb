```python
import pandas as pd
import calendar
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('extracted_info.csv')
df.head()
```

```python
df['DecisionDay'] = df['Decision Date'].apply(lambda x: x.split(' ')[0])
df['DecisionMonth'] = df['Decision Date'].apply(lambda x: list(calendar.month_name).index(x.split(' ')[1]))
df['DecisionYear'] = df['Decision Date'].apply(lambda x: x.split(' ')[2])
df['LevelOfCourt'] = df['Tribunal/Court'].apply(lambda x: 1 if x=='Court of Appeal' else (0 if x =='High Court' else -1))
df['Target'] = df['Outcome'].apply(lambda x: 1 if (x == 'Appeal allowed.' or x == 'Order accordingly.') else (-1 if x == 'Appeal dismissed' else 0))
```
